```python
import json
from datetime import datetime
import string as s
from random import *
```

```python
# Open json files
with open("chemist_store.json", "r") as acess_file:
  store_data = json.load(acess_file)

with open("Sells_Record.json", "r") as acess_json:
  sell_data = json.load(acess_json)


total_amt = []
sell = {}

# Take Input
customer_name = input("Type Your Name: ")
cs_name = customer_name
for product_num in range(0, int(input("Number of Products: "))):
  pr_id = '978735'+ str(input('Product ID of the Medicine: '))
  if len(pr_id) != 10:
    print("Enter a Valid Id of [0000]")
    pr_id = '978735' + str(input('Product ID of the Medicine: '))
  else:
    pass
  qn = int(input("Quantity being Purchased: "))

  amt = float(store_data[pr_id]["Price"])

  # Update store
  new_qn = str(int(store_data[pr_id]['Quantity']) + qn)
  store_data[pr_id]['Quantity'] = new_qn
  store_data[pr_id]['Price'] = str(amt)
  total_amt.append(float(qn) * amt)


  med_name = store_data[pr_id]["Name"]
  sell[pr_id] = [med_name, qn, amt]

sum_amt = float(sum(total_amt))

# Discount
if sum_amt >= 500:
    if sum_amt >= 9500:
        disc = (sum_amt * 0.2)  # 20%
    elif sum_amt >= 5000:
        disc = (sum_amt * 0.12) # 12%
    elif sum_amt >= 3000:
        disc = (sum_amt * 0.1)  # 10%
    elif sum_amt >= 1000:
        disc = (sum_amt * 0.05) # 5%
    elif sum_amt >= 500:
        disc = (sum_amt * 0.03) # 3%
  
elif sum_amt < 500:
    disc = (sum_amt * 0.02)   # 2%

disc = round(disc,2)
net_amt = round((sum_amt - disc),2)

# Date and time
now = datetime.now()
date = now.strftime("%d/%m/%Y %H:%M:%S")

# Receipt Number
ch = s.ascii_letters + s.digits
rcp = "".join(choice(ch) for x in range(randint(8,10)))
rcp_num = "978735-" + rcp

sell_data[rcp_num] = [cs_name, [sell], date, net_amt]

# Update Json files
with open('chemist_store.json', 'w') as js_f:
    json.dump(store_data, js_f)

with open('Sells_Record.json', 'w') as js_file:
    json.dump(sell_data, js_file)


# Billing:
print('--------------------------------------------------------------------------')
print("RECEIPT")
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("Receipt No.:   ", rcp_num)
print("Date:          ", date)
print("Customer Name: ", customer_name)
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("Medicine ID: [Med. Name, Quantity, Amt, Tot-₹]")
for key in sell:
    print(key, ":", sell[key])
print("Total Amount:   ", str(sum_amt),"₹")
print("Discount:       ", str(disc))
print("Net Amount:     ", str(net_amt),"₹")
print("Get Well Soon!!!")
print("-------------------------------------------------------------------------")
```
